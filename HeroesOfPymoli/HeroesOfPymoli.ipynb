## Age Demographics

* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], age_bins, labels=age_groups)
purchase_data_df.head()
```

```python
purchase_data_df_dropped = purchase_data_df.drop_duplicates(["SN"])
purchase_data_df_dropped.head()
```

```python
age_range_count = purchase_data_df_dropped["Age Ranges"].value_counts()
age_range_count
```

```python
pct_players = age_range_count/purchase_data_df_dropped["SN"].count() * 100
pct_players
```

```python
age_demographic_df = pd.DataFrame({"Percentage of Players" : pct_players, "Total Count" : age_range_count})
age_demographic_df.sort_index().round(2)

age_demographic_df["Percentage of Players"] = age_demographic_df["Percentage of Players"].map('{:,.2f}%'.format)
age_demographic_df.sort_index()
```
